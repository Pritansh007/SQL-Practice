SQL Practice- Introductory Problems

Q1. We have a table called Shippers. return all the fields from all the shippers

```sql
SELECT * from Shippers
```

Q2. Certain fields From Categories

```sql
SELECT CategoryName, Description
From Categories
```

Q3. Sales Representatives

```sql
SELECT FirstName, LastName, HireDate 
From Employees 
Where Title = 'Sales Representative' 
```

Q4. Sales Representatives in United States

```sql
SELECT FirstName, LastName, HireDate 
From Employees 
Where Title = 'Sales Representative' and Country ='USA' 
```

Q5. Orders Placed by Specific employeeID

```sql
SELECT OrderID, OrderDate from Orders
Where EmployeeID = 5
```

Q6. Suppliers and Contact Titles

```sql
SELECT SupplierID, ContactName, ContactTitle 
From Suppliers
Where ContactTitle <> 'Marketing Manager'
```

Q7. Products with "queso" in ProductName

```sql
Select ProductID, ProductName
From Products
Where ProductName like '%queso%'
```

Q8. Orders shipping to France or Belgium

```sql
SELECT OrderID, CustomerID, ShipCountry 
From Orders
Where ShipCountry = 'France' or 
ShipCountry = 'Belgium' 
```

Q9. Orders Shipping to any country in Latin America

```sql
SELECT OrderID, CustomerID, ShipCountry 
From Orders
Where ShipCountry in ('Brazil', 'Mexico','Argentina','Venezuela')
```

Q10. Employees, in order of age

```sql
Select FirstName, LastName, Title, BirthDate
From Employees
Order by BirthDate
```

Q11. Showing only the date with a DateTime Field

```sql
Select FirstName, LastName, Title, DateonlyBirthDate= CONVERT(Date, BirthDate)
From Employees
Order by BirthDate
```

Q12. Employees full name

```sql
Select FirstName, LastName, FullName = FirstName+' '+ LastName
From Employees
```

Q13. OrderDetails amount per line item

```sql
SELECT OrderID, ProductID, UnitPrice, Quantity, TotalPrice = UnitPrice*Quantity
From [Order Details] 
Order by OrderID, ProductID
```

Q14. How many customers?

```sql
Select TotalCustomers = Count(*)
From Customers
```

Q15. When was first order?

```sql
Select FirstOrder = min(OrderDate)
From Orders
```

Q16. Countries where there are customers

```sql
Select Country 
From Customers
Group by Country
```

Q17. Contact titles for customers

```sql
Select ContactTitle, TotalContactTitle = Count(*) 
From Customers
Group by ContactTitle
Order by count(*) desc
```

Q18. Products with associated suppliers names

```sql
SELECT ProductID, ProductName, Supplier= CompanyName
From Products
Join Suppliers on Products.SupplierID = Suppliers.SupplierID
```

Q19. Orders and the shipper that was used

```sql
Select OrderID, OrderDate = convert(date, OrderDate), Shipper= CompanyName
From Orders
Join Shippers on Shippers.ShipperID= Orders.ShipVia
Where OrderID< 10270
Order by OrderID
```
